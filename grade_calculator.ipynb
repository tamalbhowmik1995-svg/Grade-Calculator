```python
#show total, percentage and grade
import pandas as pd
df=pd.read_csv('students_marks.csv')
print(df)
def calculator(marks):
    if marks>=90:
        return 'A+'
    elif marks>=80:
        return 'A'
    elif marks>=70:
        return 'B'
    elif marks>=60:
        return 'C'
    else:
        return 'D'
df['Total']=df[['Math', 'Science', 'English']].sum(axis=1)
df['Percentage']=df['Total']/3
df['Grade']=df['Percentage'].apply(calculator)
print(df)
```

```python
#show highest and lowest percentage
highest=df['Percentage'].max()
topper=df.loc[df['Percentage']==highest, 'Name'].values[0]
lowest=df['Percentage'].min()
lower=df.loc[df['Percentage']==lowest, 'Name'].values[0]
print(f'Highest Percentage: {highest} ({topper})')
print(f'Lowest Percentage: {lowest} ({lower})')
```

```python
#showing percentage using bar chart.
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('students_marks.csv')
def calculator(marks):
    if marks>=90:
        return 'A+'
    elif marks>=80:
        return 'A'
    elif marks>=70:
        return 'B'
    elif marks>=60:
        return 'C'
    else:
        return 'D'
df['Total']=df[['Math', 'Science', 'English']].sum(axis=1)
df['Percentage']=df['Total']/3
df['Grade']=df['Percentage'].apply(calculator)
print(df)
highest=df['Percentage'].max()
lowest=df['Percentage'].min()
average=df['Percentage'].mean()
toper=df.loc[df['Percentage']<60].shape[0]
fail_count=df[df['Percentage']<60].shape[0]
print(f'Highest Percentage: {highest} ({topper})')
print(f'Highest Percentage: {lowest}')
print(f'Average Percentage: {average:.2f}')
print(f'Number of Students Failed(<60%):{fail_count}')
df.to_csv('students_report.csv', index=False)
plt.figure(figsize=(4,5))
plt.bar(df['Name'],df['Percentage'], color='blue')
plt.xlabel('Student')
plt.ylabel('Percentage')
plt.title('Student Performance')
plt.show()
```

```python
#show grade using pie chart
import matplotlib.pyplot as pl
plt.pie(df['Grade'].value_counts(), labels=df['Grade'].value_counts().index, autopct='%1.1f%%')
plt.title('Grade Distribution')
plt.show()
```
